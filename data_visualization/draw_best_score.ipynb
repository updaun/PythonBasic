```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

테마 변경

```python
plt.style.use('classic')
```

data dictionary 생성

```python
# ai stage 리더보드를 날짜순으로 정렬 후 복사하여 txt파일을 만들어 주세요.
with open("data/submission3.txt", "r", encoding="utf-8") as my_file:
    contents_list = my_file.readlines()

data_dict = dict()

for i in contents_list:
    data_dict[i.split('\t')[3][5:10]] = []
for i in contents_list:
    data_dict[i.split('\t')[3][5:10]].append(float(i.split('\t')[1]))

print(data_dict)
```

best score 생성

```python
scores = []

for i in sorted(data_dict.keys()):
    print( i , max(data_dict[i]))
    scores.append(max(data_dict[i]))
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,5))
fig.patch.set_facecolor('xkcd:gray')

np.random.seed(97)
x = range(0, len(scores))
y = scores

ax.plot(x, y,
        color='black',
        marker='*',
        linestyle='--',
        markersize=15,
        markerfacecolor='yellow',
        )
plt.yticks(np.arange(0.3, 1.0, 0.1))
plt.xticks(np.arange(0, len(data_dict.keys())), labels=sorted(data_dict.keys()))
plt.title('Best LB Score of Day')
plt.xlabel('Date')
plt.ylabel('mIoU')

plt.show()
```
