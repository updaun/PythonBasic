### Line plot
- 연속적으로 변화하는 값을 순서대로 점으로 나타내고, 이를 선으로 연결한 그래프

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
with open("data/submissions.txt", "r", encoding="utf-8") as my_file:
    contents_list = my_file.readlines()
    
scores = [float(i.split('\t')[1]) for i in contents_list]
# scores = scores[::-1]
len(scores)
# scores
```

```python
plt.style.use('classic')
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,5))

np.random.seed(97)
x = range(0, len(scores))
y = [scores[i] for i in x]

x = range(0, len(scores)+1)
y.insert(0,0)

ax.plot(x, y,
        color='black',
        marker='o',
        linestyle='--'
        )
plt.yticks(np.arange(0, 0.9, 0.1))
plt.xticks(np.arange(0, 20), labels=[
                                '',
                                '03-21',
                                '03-22',
                                '03-23',
                                '03-24',
                                '03-25',
                                '03-26',
                                '03-27',
                                '03-28',
                                '03-29',
                                '03-30',
                                '03-31',
                                '04-01',
                                '04-02',
                                '04-03',
                                '04-04',
                                '04-05',
                                '04-06',
                                '04-07',
                                ""])
plt.title('Best LB Score of Day')
plt.xlabel('Date')
plt.ylabel('mAP')
plt.show()
```
