##  EDA(Exploratory Data Analysis)

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/young_survey.csv')
df.head()
```

```python
 df.shape
```

## 기본 정보 출력

```python
basic_info = df.iloc[:, 140:]
basic_info.head()
```

```python
basic_info.describe()
```

```python
basic_info['Gender'].value_counts()
```

```python
basic_info['Handedness'].value_counts()
```

```python
basic_info['Education'].value_counts()
```

```python
sns.violinplot(data=basic_info, y='Age')
```

```python
sns.violinplot(data=basic_info, x='Gender', y='Age')
```

```python
sns.violinplot(data=basic_info, x='Gender', y='Age', hue='Handedness')
```

```python
sns.jointplot(data=basic_info, x='Height', y='Weight')
```

## 요즘 인기있는 직업은?

```python
o_df = pd.read_csv('data/occupations.csv')
o_df
```

```python
female_o_df = o_df[o_df['gender'] == 'F']
female_o_df
```

```python
female_o_df['occupation'].value_counts()
```

```python
female_o_df['occupation'].value_counts().sort_values(ascending=False)
```

```python
man = o_df[o_df['gender']=='M']
man
```

```python
man['occupation'].value_counts().sort_values(ascending=False)
```

## 클러스터 분석(Cluster Analysis)

```python
interests = df.loc[:,'History':'Pets']
interests.head()
```

```python
corr = interests.corr()
corr
```

```python
corr['History'].sort_values(ascending=False)
```

```python
sns.clustermap(corr)
```
