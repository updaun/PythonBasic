```python
def solution(n):
    num_list = []
    i = 1
    if n == 1:
        return 0
    else:
        for i in range(1,n**2):
            if i < n**2: 
                if i // n == i % n:
                    num_list.append(i)
                i += 1
        return sum(num_list)
```

```python
print(solution(2))
```
