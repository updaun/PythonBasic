# 표기법과 개념
- 시간복잡도
- 공간복잡도
- 점근 표기법(Big-O)

```python
def solution(N):
    output_string = ''
    for i in range(1, N+1):
        if N % i == 0:
            if i != N:
                output_string += str(i)+' '
            else:
                output_string += str(i)
    print(output_string)
```

```python
solution(6)
```

```python
solution(6000000)
```

## 풀이
- 시간복잡도는 보통 big-o로 표현한다.
- big-o 를 생각했을 때 현재의 방법은 O(N)으로 효율성이 낮다.
- 가우스와 비슷한 방법으로 좌우 대칭을 고려하면 효율성을 높일 수 있다.

```python
n_list = list(map(int, input().split()))

max = n_list[0]
min = n_list[0]

def list_max():
    for i in range(len(n_list)):
        s = 0
        for j in range(len(n_list)):
            if n_list[i] < n_list[j]:
                s += 1
        if s == 0:
            return i
        
def list_min():
    for i in range(len(n_list)):
        s = 0
        for j in range(len(n_list)):
            if n_list[i] > n_list[j]:
                s += 1
        if s == 0:
            return i
    
print(n_list[list_max()], n_list[list_min()])
```

```python
n_list = list(map(int, input().split()))

for i in range(len(n_list)):
    s = 0
    k = 0
    for j in range(len(n_list)):
        if n_list[i] < n_list[j]:
            s += 1
        if n_list[i] > n_list[j]:
            k += 1
    if s == 0:
        max_idx = i
    if k == 0:
        min_idx = i

print(n_list[max_idx], n_list[min_idx])
```
