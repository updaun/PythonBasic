```python
stack_list = []
def process_stack(command):
    global res
    if command[0] == 'push':
        stack_list.append(int(command[1]))
        
    elif command[0] == 'pop':
        if len(stack_list) != 0:
            res = int(stack_list.pop())
        else:
            res = -1
        print(res)

    elif command[0] == 'size':
        res = len(stack_list)
        print(res)
    
    elif command[0] == 'empty':
        if len(stack_list) == 0:
            res = 1
        else:
            res = 0 
        print(res)
    elif command[0] == 'top':
        if len(stack_list) == 0:
            res = -1
        else:
            res = int(stack_list[-1])
        print(res)
    
            
n = int(input("n = "))
for i in range(n):
    command = input("command = ").split(" ")
    process_stack(command)
```
