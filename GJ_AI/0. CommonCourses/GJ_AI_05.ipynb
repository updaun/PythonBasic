## 지불 금액, 팁과의 상관관계를 산점도로 도시

```python
scatter_plot = plt.figure()
plt.figure()

axes1 = scatter_plot.add_subplot(1,1,1) # 그래프 1개이므로 굳이 안해도 되는 부분 

axes1.scatter(tips['total_bill'],
                   tips['tip'],
                  color = 'red',
                  label = 'tip and total_bill',
                  alpha = 0.5) # 투명도를 주면 점이 모여도 확인하기 용이하다.

axes1.set_title("Scatter_Plot")
axes1.set_xlabel("Total_bill")
axes1.set_ylabel("Tip")
axes1.legend()

axes1.grid()
```

### 40~50 달러 확대 분석

```python
total_bill_row = []
tip_row = []

for i in range(len(tips['total_bill'])):
    if tips['total_bill'][i] >= 40:
        total_bill_row.append(tips['total_bill'][i])
    if tips['total_bill'][i] >= 40:
        tip_row.append(tips['tip'][i])
        
print("total_bill_row : ", total_bill_row)
print("tip_row : ", tip_row)
```

```python
plt.plot(total_bill_row, tip_row, 'o')

plt.suptitle('Tips Dataset(total_bill >= 40)')

plt.xlabel("Total_bill")
plt.ylabel("Tip")

plt.grid()
```

```python
ax = sns.regplot(x=total_bill_row, y=tip_row )
ax.grid()
```

```python
ax = plt.subplots()
ax = sns.regplot(x = tips['total_bill'], y = tips['tip'])

ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.grid()
```

## box plot

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
              labels = ['Female', 'Male'])
               
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')

axes1.set_title('Boxplot of Tips by sex')
plt.show()
```

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['day']=='Thur']['total_bill'],
               tips[tips['day']=='Fri']['total_bill'],
               tips[tips['day']=='Sat']['total_bill'],
               tips[tips['day']=='Sun']['total_bill']],
              labels = ['Thur', 'Fri', 'Sat', 'Sun'])
               
axes1.set_xlabel('Day')
axes1.set_ylabel('Total_bill')

axes1.set_title('Boxplot of Total_bill by Day')
plt.show()
```

### 추가적인 변수를 그리는 방법(아이디어)
- 성별을 추가적으로 반영 -> 점의 색으로 구분
- 같이 방문한 인원의 수 -> 점의 크기

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

## apply 메서드

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips[['sex', 'sex_color']].head(8)
```

```python
plt.figure(figsize = (40,5))
plt.title("Ebola dataset", fontsize = 50)

plt.plot(ebola['Date'], ebola['Cases_Liberia'], 'o', color='red',
        label = 'Cases_Liberia_count')

plt.plot(ebola['Date'], ebola['Deaths_Liberia']), 'o', color='black',
        label = 'Deaths_Ligeria_count'

plt.axhline(y = ebola['Cases_Liberia'].mean(), 
            linestyle = '--',
            color = 'red',
            label = 'Cases_Liberia_mean')

plt.axvline(x = ebola['Date'][61],
           color = 'blue',
           label = 'line')

plt.xticks(rotation = 90, size = 20)
plt.yticks(size = 20)

plt.xlabel('Date', size = 40)
plt.ylabel("Count of Cases_Liberia", size = 40)
plt.legend(fontsize = 20)
plt.show()
```
