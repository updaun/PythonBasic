```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7,7,0.1)
f_x = sigmoid(z)
```

```python
# 데이터 시각화
plt.plot(z,f_x)
plt.axvline(0.0 , color = 'k')
plt.ylim(-0.1,1.1)
plt.xlabel('z')
plt.ylabel('f(x)')

plt.yticks([0.0,0.5,1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```
