## 데이터 준비

```python
fish = pd.read_csv('data/Fish.csv')
fish.head()
```

```python
fish['Species'].value_counts()
```

```python
df = fish[['Species', 'Weight', 'Length2']]
df
```

```python
# Length2 -> Length
df.columns = ['Species', 'Weight', 'Length']
df.head()
```

```python
filter = df['Species'].isin(['Bream', 'Smelt'])
df = df.loc[filter]
df
```

```python
def integer_encoding(x):
    if x == 'Bream':
        return 1
    else:
        return 0
```

```python
# 분류하고자 하는 대상을 '1'로 지정하는 것이 좋다.
df['class'] = df['Species'].apply(integer_encoding)
df
```

```python
df = df.reset_index()
df.drop('index', axis = 1, inplace = True)
df
```

```python
df.to_csv('data/fish-data.csv', index =False)
```
